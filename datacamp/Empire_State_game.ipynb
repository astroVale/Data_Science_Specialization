```python
# Initialization
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  

np.random.seed(123)
```

```python
# Initialize all_walks
all_walks = []

# Simulate random walk 1000 times
for i in range(1000) :

    # Code from empire_game.py
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
            
        # Implement clumsiness
        if np.random.rand() <= 0.001 :
            step = 0
            
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Convert all_walks to Numpy array and transpose (otherwise python misunderstands): np_aw_t 
np_aw_t = np.transpose(np.array(all_walks))
```

```python
# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

```python
# Plot histogram of the end points of all the random walks:
# Select last row from np_aw_t: ends
ends = np_aw_t[-1]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

```python
# Compute the chance to reach 60 steps high if playing the Empire State Building game --> 78.80 %
p_win = np.count_nonzero(ends[ends >= 60]) / 1000.
print(str(p_win * 100) + " %")
```
